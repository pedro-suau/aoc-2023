```python
# Helper function to get the first and last digit of a string
def get_first_and_last_digit(s):
  i=0
  j= len(s) - 1
  while i < len(s):
    if s[i].isdigit():
      first_digit = s[i]
      break
    else:
      i += 1
  while j >= 0:
    if s[j].isdigit():
      last_digit = s[j]
      break
    else:
      j -= 1
  return int(str(first_digit) + str(last_digit))
    
```

```python
acc=0
for i in range(len(puzzle_data)):
  acc += get_first_and_last_digit(puzzle_data[i])
acc
```

```python
# Part 2
# Dict that maps a number string to its value
import re
number_dict = {'one': 1, 'two': 2, 'three': 3, 'four': 4,
               'five': 5, 'six': 6, 'seven': 7, 'eight': 8, 'nine': 9}

def get_digit_from_string(s):
  number_index_dict = {}
  for number in number_dict.keys():
    if number in s:
       number_index_dict[number] = s.find(number)
  sorted_dict = dict(sorted(number_index_dict.items(), key=lambda item: item[1]))
  return sorted_dict

def replace_number_with_digit(s):
  formatted_string = s
  array_of_numbers= get_digit_from_string(s)
  for number in array_of_numbers:
    if number in s:
      formatted_string = formatted_string.replace(number, str(number_dict[number])+number[-1])
  return formatted_string



mock_list = ['two1nine',
'eightwothree',
'abcone2threexyz',
'xtwone3four',
'4nineeightseven2',
'zoneight234',
'7pqrstsixteen']
acc3= 0
for i in range(len(puzzle_data)):
  acc3 += get_first_and_last_digit(replace_number_with_digit(puzzle_data[i]))
acc3

# acc2=0
# for i in range(len(mock_list)):
#   acc2 += get_first_and_last_digit(replace_number_with_digit(mock_list[i]))
# acc2
```
